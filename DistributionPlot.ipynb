#DISTRIBUTION PLOTS

```python
#are for quantative data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# seaborn mai already data sets hote hai
df = sns.load_dataset("tips")
df
```

DISTPLOT / HISTPLOT -> smae

```python
df['size'].unique() #to see categorical data
```

```python
sns.displot(df['total_bill']) #to create distribution plot -> sns.displot(var name/data)
```

```python
#bins -> the dabbe
#ye jo dabbe bante hai graph pe inhe hum bins bolte hai (ienka bin size hota hai and number of bina )
#values according to bins size mai padhta hai
#kde(kernal density estimate) ye line dalta hai to tell us normal hai kya distribution ya left ki taraf ya right ki taraf
```

```python
sns.displot(df['total_bill'],bins = 100) #can decide size of your bins
sns.histplot(df['total_bill'],bins = 100,kde = True) #dist ya hist -  same hi hai
```

```python
#ik canvas pai do graph
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins = 25 , kde = True)
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins = 25 , kde = True)
```

#JOINTPLOT

```python
sns.jointplot(x = 'total_bill' , y = 'tip' ,data = df, kind = 'scatter')
#scatter = linear reg mai imp
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' ,data = df, kind = 'kde')
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' ,data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'total_bill' , y = 'tip' ,data = df, kind = 'reg')
```
