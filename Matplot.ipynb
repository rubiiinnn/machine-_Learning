```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11) #linspace linear chlta , o se 5 tk values print inke beech mai 11 ki spacing
y = 2*x
x , y
```

```python
plt.plot(x,y) #line graph plot hoga
plt.title("TITLE")
plt.xlabel("x label")
plt.ylabel("y label")
```

```python
#SUBPLOTS - ik saath multiple plots krna
```

```python
#plt.sublpot(rows,column , konsa plot slect kr rhe ho)
# White backgroun -> canvas - uske upr graph - 2*2 : 4 graph , 3*3 :9 graph
plt.subplot(2,2,4)
plt.plot(x,y) # 4 wale graph mai plotting ho jayegi

plt.subplot(2,2,3)
plt.plot(x**5,y)

plt.subplot(2,2,2)
plt.plot(x,y**5)

plt.subplot(2,2,1)
plt.plot(x**4,y)
```

```python
#OBJECT ORIENTED WAY
#canvas ko access - uspe chnages
```

```python
fig = plt.figure(figsize=(20,8),dpi = 100) #fig ik var h #DPI = DEPTH PER INCH : pixels ko control krna
axis1 = fig.add_axes([0.1 , 0.1 , 1 , 1]) #fig.add_axes(left se shift , bottom se shift , widht , height){values 0<=1 tk}
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.5,0.3])
axis2.plot(x,y)
```

**TYPES OF PLOTS**

2) Histogram

```python
#random module se sample function import kiya.
from random import sample
#sample = ek function hai jo random unique elements choose karta hai.
#1 se 999 tak me se 100 random unique numbers liye
# unko data me store kiya.
data = sample(range(1,1000),100)
plt.hist(data)
```

3) Boxplot

```python
plt.boxplot(x)
```
