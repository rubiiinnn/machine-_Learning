```python
#LINEAR REGRESSION
```

```python
import numpy as np
import matplotlib.pyplot as plt #to plot
```

```python
#Training data
x_train = np.array([1.0,2.0])
y_train = np.array([300.0 , 500.0])
```

```python
#Computing cost
def compute_cost(x,y,w,b):
  m = x.shape[0] #no of training exmaple , here m = 2
  cost_sum = 0 #initial cost
  for i in range(m):
    f_wb = w*x[i] + b
    cost = (f_wb - y[i]) ** 2 #predicted - actual
    cost_sum = cost_sum + cost
  total_cost = (1/(2*m)) * cost_sum #avg error

  return total_cost
```

```python
#Testing with data
```

```python
w = 100
b = 200
cost = compute_cost(x_train , y_train , w , b)
print(cost)
```

```python
#plotting
plt.scatter(x_train, y_train, marker='x' , c = 'r')
plt.xlabel("House size(1000 sq ft)")
plt.ylabel("Price(1000s of $)")
plt.title("Training Data")
plt.show()
```
